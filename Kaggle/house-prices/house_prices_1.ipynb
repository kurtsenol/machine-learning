```
df[["BsmtHalfBath", "BsmtFullBath", "BsmtFinSF1", "GarageCars", "GarageArea", "TotalBsmtSF", "BsmtUnfSF", "BsmtFinSF2"]]=df[["BsmtHalfBath", "BsmtFullBath", "BsmtFinSF1", "GarageCars", "GarageArea", "TotalBsmtSF", "BsmtUnfSF", "BsmtFinSF2"]].fillna(0)
```

```
train_df = pd.concat([train, labels], axis=1)
train_df
```

```
train = train_df.sample(frac=0.7,random_state=21)
eval = train_df.drop(train.index)
```

```
eval
```

```
X_train = train.drop("SalePrice", axis=1).values
y_train = train.SalePrice.values

X_eval=eval.drop("SalePrice", axis=1).values
y_eval = eval.SalePrice.values
```

```
from sklearn.metrics import mean_squared_log_error
from sklearn.ensemble import GradientBoostingRegressor


grad_boost = GradientBoostingRegressor(random_state=21, n_estimators=1000)
grad_boost.fit(X_train,y_train)

acc_train_gb = np.sqrt(mean_squared_log_error(y_train, grad_boost.predict(X_train)))
acc_eval_gb = np.sqrt(mean_squared_log_error(y_eval, grad_boost.predict(X_eval)))

print (f'GB: Train = {acc_train_gb}  Evaluate = {acc_eval_gb}')
```

```
X_full = train_df.drop("SalePrice", axis=1).values
y_full = train_df.SalePrice.values
X_test = test.values
```
