```python
import pandas as pd
import numpy as np
import spacy
import os
import warnings
import pickle 
warnings.simplefilter("ignore")
```

```python
import plotly.express as px
import plotly.offline as pyo
import plotly.graph_objs as go
from plotly.subplots import make_subplots
import plotly.figure_factory as ff

pyo.init_notebook_mode(connected=True)
```

```python
def configure_plotly_browser_state():
  import IPython
  display(IPython.core.display.HTML('''
        <script src="/static/components/requirejs/require.js"></script>
        <script>
          requirejs.config({
            paths: {
              base: '/static/base',
              plotly: 'https://cdn.plot.ly/plotly-latest.min.js?noext',
            },
          });
        </script>
        '''))
```

```python
path = "/content/drive/MyDrive/Colab Notebooks/TopicModeling/Methodology/data/Methodology_sections/"
```

```python
files = os.listdir(path)
files
```

```python
m1 = pd.read_csv(path + files[0])
```

```python
m1
```

```python
section1 = m1.Method.iloc[0]
section1
```

```python
sentences = []
nlp = spacy.load("en_core_web_sm")
doc = nlp(section1)
for sent in doc.sents:
  sentences.append(sent.text)
  sentences = [s for s in sentences if len(s)>20]
```

```python
sentences
```

```python
nlp_model_last = spacy.load("/content/drive/MyDrive/Colab Notebooks/TopicModeling/Methodology/data/prodigy_first_model/model-last")
```

```python
for sent in sentences:
  doc = nlp_model_last(sent)
  spacy.displacy.render(doc, style="ent", jupyter=True) # display in Jupyter
```

```python
doc = nlp_model_last(section1)
spacy.displacy.render(doc, style="ent", jupyter=True) 
```

```python
sections = m1.loc[m1.Method.notna(), "Method"].tolist()
sections 
```

```python
for section in sections:
  doc = nlp_model_last(section)
  spacy.displacy.render(doc, style="ent", jupyter=True) 
  print("\n" + "-"*30 + "\n")
```

```python
df_m = pd.read_pickle("/content/drive/MyDrive/Colab Notebooks/TopicModeling/Methodology/df_method_filtered.pkl")
df_m
```

```python
np.random.seed(0)
int_arr = np.random.randint(0,2775, 20)
int_arr
```

```python
for i in int_arr:
  
  print(df_m.loc[i, "Method"].upper())
  doc = nlp_model_last(df_m.loc[i, "methodology"])
  spacy.displacy.render(doc, style="ent", jupyter=True) 
  print("\n" + "-"*30 + "\n")
```

```python
res = []
for i in range(len(df_m)):
  method = df_m.loc[i, "Method"].upper()
  doc = nlp_model_last(df_m.loc[i, "methodology"])
  ents = [e.text for e in doc.ents]
  res.append((i,method, ents))
```

```python
with open('/content/drive/MyDrive/Colab Notebooks/TopicModeling/Methodology/data/prodigy_results_last.pkl', 'wb') as f:
  pickle.dump(res, f)
```

```python
nlp_model_best = spacy.load("/content/drive/MyDrive/Colab Notebooks/TopicModeling/Methodology/data/prodigy_first_model/model-best")
```

```python
res_best = []
for i in range(len(df_m)):
  method = df_m.loc[i, "Method"].upper()
  doc = nlp_model_best(df_m.loc[i, "methodology"])
  ents = [e.text for e in doc.ents]
  res_best.append((i,method, ents))
```

```python
with open('/content/drive/MyDrive/Colab Notebooks/TopicModeling/Methodology/data/prodigy_results_best.pkl', 'wb') as f:
  pickle.dump(res_best, f)
```

```python
with open('/content/drive/MyDrive/Colab Notebooks/TopicModeling/Methodology/data/prodigy_results_last.pkl', 'rb') as f:
  res_last = pickle.load(f)
```

```python
df_last = pd.DataFrame(res_last)
df_last.columns = ["Document", "Methodology", "Entities"]
df_last["Entity_Count"] = df_last["Entities"].apply(len)
df_last["Unique_Entity_Count"] = df_last["Entities"].apply(set).apply(len)
df_last
```

```python
df_last.sum()
```

```python
configure_plotly_browser_state()

fig = px.box(df_last, y = "Entity_Count", x = "Methodology", hover_data=["Document"])
fig.update_layout(
    autosize=False,
    width=1500,
    height=600,
    title_text=f"Model_Last Entity Counts Per Methodology",

)
fig.show()
```

```python
configure_plotly_browser_state()

fig = px.box(df_last, y = "Unique_Entity_Count", x = "Methodology", hover_data=["Document"])
fig.update_layout(
    autosize=False,
    width=1500,
    height=600,
    title_text=f"Model_Last Unique Entity Counts Per Methodology",

)
fig.show()
```
